```python
import numpy as np
import pandas as pd
```

```python
data = pd.Series([3,6,9,12,np.nan,15,18,21,np.nan])
data
```

```python
data.notnull()
```

```python
dict1 = {
    0:[3,6,9,12,np.nan],
    1:[15,18,21,np.nan,np.nan],
    2:[24,np.nan,np.nan,np.nan,np.nan],
    3:[np.nan,np.nan,np.nan,np.nan,np.nan],
    4:[27,np.nan,30,np.nan,33]
}
```

```python
df = pd.DataFrame(dict1)
df
```

```python
df.dropna(axis=1,how='all')
```

```python
df_mean=df.fillna(df.mean())
df_mean
```

```python
df_median=df.fillna(df.median())
df_median
```

```python
df_mode=df.fillna(df.mode())
df_mode
```

###There is no value that occur more than one time so fill with mode will be a null value.
